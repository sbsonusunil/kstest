# KS-Test
the test is used to perform the probability distribution between 2 sample set follows same distribution or not

```python
from scipy import stats
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
x=stats.norm.rvs(size=1000)
sns.set_style('whitegrid')
sns.kdeplot(np.array(x),bw_method=0.5)
plt.show()
```

```python
stats.kstest(x,'norm')
```

we accept null hypothesis as t value is too small than p value


```python
#continuos uniform distribtuion
y=np.random.uniform(0,1,10000)
sns.set_style('whitegrid')
sns.kdeplot(np.array(y),bw_method=0.1)
plt.show()
```

```python
stats.kstest(y,'norm')
```

we reject null hypothesis as pvalue is too small to compare with t value
